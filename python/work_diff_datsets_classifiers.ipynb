```python
import numpy as np
import os
import pandas as pd

from PIL import Image
import tqdm
import copy

import torchvision.transforms.functional as TF
from torchvision.transforms import ToTensor, RandomHorizontalFlip, RandomVerticalFlip
import torch.nn as nn

import torchvision.models as models 
import torch 
```

```python
df_train_1 = pd.read_csv("/content/drive/MyDrive/train-vgg19_fc1.csv")
df_test_1 = pd.read_csv("/content/drive/MyDrive/test-vgg19_fc1.csv")
df_train_2 = pd.read_csv("/content/drive/MyDrive/train-vgg19_fc2.csv")
df_test_2 = pd.read_csv("/content/drive/MyDrive/test-vgg19_fc2.csv")
df_train_3 = pd.read_csv("/content/drive/MyDrive/train-vgg19_fc3.csv")
df_test_3 = pd.read_csv("/content/drive/MyDrive/test-vgg19_fc3.csv")
df_train_3 = pd.read_csv("/content/drive/MyDrive/train4.csv")
df_test_3 = pd.read_csv("/content/drive/MyDrive/test4.csv")
```

```python
import matplotlib.pyplot as plt

from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import SGDClassifier
```

```python
x_train_1 = df_train_1.iloc[:, 2:1002]
y_train_1 = df_train_1.loc[:, "labels"]

x_test_1 = df_test_1.iloc[:, 2:1002]
y_test_1 = df_test_1.loc[:, "labels"]

x_train_2 = df_train_2.iloc[:, 2:4098]
y_train_2 = df_train_2.loc[:, "labels"]

x_test_2 = df_test_2.iloc[:, 2:4098]
y_test_2 = df_test_2.loc[:, "labels"]

x_train_3 = df_train_3.iloc[:, 2:4098]
y_train_3 = df_train_3.loc[:, "labels"]

x_test_3 = df_test_3.iloc[:, 2:4098]
y_test_3 = df_test_3.loc[:, "labels"]

x_train_4 = df_train_3.iloc[:, 2:4098]
x_train_4.join(x_train_1) 
y_train_4 = df_train_3.loc[:, "labels"]

x_test_4 = df_test_3.iloc[:, 2:4098]
x_test_4.join(x_test_1) 
y_test_4 = df_test_3.loc[:, "labels"]

x_train_5 = df_train_3.iloc[:, 2:4098]
x_train_5.join(x_train_1)
x_train_5.join(x_train_2) 
y_train_5 = df_train_3.loc[:, "labels"]

x_test_5 = df_test_3.iloc[:, 2:4098]
x_test_5.join(x_test_1)
x_test_5.join(x_test_2) 
y_test_5 = df_test_3.loc[:, "labels"]

x_train_6 = df_train_3.iloc[:, 2:4098]
x_train_6.join(x_train_2) 
y_train_6 = df_train_3.loc[:, "labels"]

x_test_6 = df_test_3.iloc[:, 2:4098]
x_test_6.join(x_test_2) 
y_test_6 = df_test_3.loc[:, "labels"]

```

```python
for i in range(1, 25):
  print("-------" + str(i) + "------")
  clf_SVC = SVC(C=i)
  clf_SVC.fit(x_train_1, y_train_1)

  print(f"1: {accuracy_score(y_test_1, clf_SVC.predict(x_test_1))*100:.2f}")


  clf_SVC = SVC(C=i)
  clf_SVC.fit(x_train_2, y_train_2)

  print(f"2: {accuracy_score(y_test_2, clf_SVC.predict(x_test_2))*100:.2f}")

  clf_SVC = SVC(C=i)
  clf_SVC.fit(x_train_3, y_train_3)

  print(f"4: {accuracy_score(y_test_3, clf_SVC.predict(x_test_3))*100:.2f}")

  clf_SVC = SVC(C=i)
  clf_SVC.fit(x_train_4, y_train_4)

  print(f"4: {accuracy_score(y_test_4, clf_SVC.predict(x_test_4))*100:.2f}")

  clf_SVC = SVC(C=i)
  clf_SVC.fit(x_train_5, y_train_5)

  print(f"5: {accuracy_score(y_test_5, clf_SVC.predict(x_test_5))*100:.2f}")

  clf_SVC = SVC(C=i)
  clf_SVC.fit(x_train_6, y_train_6)

  print(f"6: {accuracy_score(y_test_6, clf_SVC.predict(x_test_6))*100:.2f}")
```

```python
clf_LR = LogisticRegression(random_state=0, max_iter=50000)
clf_LR.fit(x_train_1, y_train_1)

print(f"1: {accuracy_score(y_test_1, clf_LR.predict(x_test_1))*100:.2f}")

clf_LR = LogisticRegression(random_state=0, max_iter=50000)
clf_LR.fit(x_train_2, y_train_2)

print(f"2: {accuracy_score(y_test_2, clf_LR.predict(x_test_2))*100:.2f}")

clf_LR = LogisticRegression(random_state=0, max_iter=50000)
clf_LR.fit(x_train_3, y_train_3)

print(f"3: {accuracy_score(y_test_3, clf_LR.predict(x_test_3))*100:.2f}")

clf_LR = LogisticRegression(random_state=0, max_iter=50000)
clf_LR.fit(x_train_4, y_train_4)

print(f"4: {accuracy_score(y_test_4, clf_LR.predict(x_test_4))*100:.2f}")

clf_LR = LogisticRegression(random_state=0, max_iter=50000)
clf_LR.fit(x_train_5, y_train_5)

print(f"5: {accuracy_score(y_test_5, clf_LR.predict(x_test_5))*100:.2f}")

clf_LR = LogisticRegression(random_state=0, max_iter=50000)
clf_LR.fit(x_train_6, y_train_6)

print(f"6: {accuracy_score(y_test_6, clf_LR.predict(x_test_6))*100:.2f}")
```
